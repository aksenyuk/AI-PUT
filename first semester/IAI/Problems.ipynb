```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
nr_elems = 200
t = np.linspace(0,10,nr_elems)
x = t
y = x +np.random.normal(0,2,nr_elems)
plt.scatter(x,y) 

x = t
y = x+10 +np.random.normal(0,2,nr_elems)
plt.scatter(x,y)
```

```python
nr_elems = 100
t = np.linspace(0,10,nr_elems)
x = t
y = x**2 -np.clip(np.random.normal(0,5,nr_elems),0,None)*10
plt.scatter(x,y) 

x = t
y = x**2+10 +np.clip(np.random.normal(0,5,nr_elems),0,None)*10
plt.scatter(x,y)
```

```python
nr_elems = 100
t = np.linspace(10,200,nr_elems)

x = np.sin(0.01*t)*t +np.random.normal(0,2,nr_elems)
y = np.cos(0.01*t)*t +np.random.normal(0,2,nr_elems)
plt.scatter(x,y)

x = np.sin(0.01*t)*(t-10)+np.random.normal(0,2,nr_elems)
y = np.cos(0.01*t)*(t-10)-10+np.random.normal(0,2,nr_elems)
plt.scatter(x,y)
```

```python
nr_elems = 500
t = np.linspace(0,20,nr_elems)

x = t
y = np.cos(t) +np.random.normal(0,0.1,nr_elems)
plt.scatter(x,y)

x = t
y = np.sin(t)+np.random.normal(0,0.1,nr_elems)
plt.scatter(x,y)
```

```python
nr_elems = 200
t = np.linspace(6,20,nr_elems)
a =0.1
b =0.1
x = a*np.exp(b*t)*np.sin(t)+np.random.normal(0,0.01,nr_elems)
y = a*np.exp(b*t)*np.cos(t)+np.random.normal(0,0.01,nr_elems)
plt.scatter(x,y)


a =0.11
b =0.11
x = a*np.exp(b*t)*np.sin(t)+np.random.normal(0,0.01,nr_elems)
y = a*np.exp(b*t)*np.cos(t)+np.random.normal(0,0.01,nr_elems)
plt.scatter(x,y)
```

```python
nr_elems=200
t = np.linspace(40,200,nr_elems)
x = np.sin(t)*t
y = np.cos(t)*t
plt.scatter(x,y)
x = np.sin(t+10)*t
y = np.cos(t+10)*t
plt.scatter(x,y)
```

```python
data = np.mgrid[0:20:0.7, 0:20:1].reshape(2,-1).T

x = data[:,0]
y = data[:,1]

z = (np.sin(x)/2 +np.sin(y)/2).round(0)
nr_elems = np.sum([z!=0])
x = x[z!=0] +np.random.normal(0,0.5,nr_elems)
y = y[z!=0] +np.random.normal(0,0.5,nr_elems)
z = z[z!=0]
plt.scatter(x,y,c=z)
```

```python
data = np.mgrid[0:20:0.1, 0:20:0.1].reshape(2,-1).T

x = data[:,0]
y = data[:,1]

z = (np.sin(x)/2 +np.sin(y)/2)

plt.scatter(x,y,c=z)
```
